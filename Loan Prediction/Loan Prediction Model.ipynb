```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
```

```python
data.columns
```

```python
data.info()
```

```python
X = data[['Gender', 'Married', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']]
y = data['Loan_Status']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("F1 Score:",metrics.f1_score(y_test, y_pred))
```

```python
def roc_curve():
    probability = logreg.predict_proba(X_test)[::,1]
    ping, swing, _ = metrics.roc_curve(y_test,  probability)
    auc = metrics.roc_auc_score(y_test, probability)
    plt.plot(ping,swing,label="data 1, auc="+str(auc))
    plt.legend(loc=4)
    plt.show()
```

```python
roc_curve()
```
