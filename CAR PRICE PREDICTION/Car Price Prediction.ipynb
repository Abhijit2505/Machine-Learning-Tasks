```python
reg_features = ['wheelbase','carlength', 'carwidth','curbweight','enginesize','horsepower','citympg','highwaympg']
reg_target = ['price']
```

```python
X = data[reg_features]
y = data[reg_target]
```

### Linear Regression 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures 
X = data[reg_features]
y = data[reg_target]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
poly = PolynomialFeatures(degree = 2) 
X_poly = poly.fit_transform(X_train) 
X_poly_test = poly.fit_transform(X_test)
  
poly.fit(X_poly, y_train) 
lin2 = LinearRegression() 
lin2.fit(X_poly, y_train) 
```

```python
y_train_predicted = lin2.predict(X_poly)
y_test_predict = lin2.predict(poly.fit_transform(X_test))
```

```python
from sklearn import metrics
print("-------Test Data--------")
print('MAE:', metrics.mean_absolute_error(y_test, y_test_predict))
print('MSE:', metrics.mean_squared_error(y_test, y_test_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_test_predict)))

print("\n-------Train Data--------")
print('MAE:', metrics.mean_absolute_error(y_train,y_train_predicted))
print('MSE:', metrics.mean_squared_error(y_train, y_train_predicted))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_train, y_train_predicted)))
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor  
  
# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 2)  
  
# fit the regressor with X and Y data 
regressor.fit(X_train, y_train) 
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics
print("-------Test Data--------")
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

### Conclusion

The **Linear Regression** model has a RMSE value of - **3420**

The **Polynomial Regression** model has a RMSE value of - **3643**

The **Decision Tree Regression** model has a RMSE value of - **2405**, with a random state of 2.

Decision Tree regression demonstrate the lowest value of RMSE, hence it's the best model to predict car price for this data.
